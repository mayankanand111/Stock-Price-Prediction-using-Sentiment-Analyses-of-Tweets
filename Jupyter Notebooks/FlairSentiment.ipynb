```python
stopwords = ["for", "on", "an", "a", "of", "and", "in", "the", "to", "from"]
```

```python
import numpy as np
import re

def clean_tweet(tweet):
    if type(tweet) == np.float:
        return ""
    temp = tweet.strip('"')
    temp = temp.lower()
    temp = re.sub("'", "", temp) # to avoid removing contractions in english
    temp = re.sub("@[A-Za-z0-9_]+","", temp)
    temp = re.sub("#[A-Za-z0-9_]+","", temp)
    temp = re.sub("$[A-Za-z0-9_]+","", temp)
    temp = re.sub(r'http\S+', '', temp)
    temp = re.sub('[()!?]', ' ', temp)
    temp = re.sub('\[.*?:\]',' ', temp)
    temp = re.sub(r'[^\x00-\x7F]+', '',temp)
    temp = re.sub("[^a-z]"," ", temp)
    temp = temp.split()
    temp = [w for w in temp if not w in stopwords]
    temp = " ".join(word for word in temp)
    return temp
```

```python
from scipy import stats
```

```python
New_Sentiment_Class=[]
for index, row in merged_data_NVDA2.iterrows():
    probs = []
    sentiments = []
    x = merged_data_NVDA2.Text3_converted[index]
    results = []
    results = [clean_tweet(tw) for tw in x]
    df1 = pd.DataFrame(results, columns =['Tweets'])
    df1['Tweets'].replace('', np.nan, inplace=True)
    df1 = df1. dropna()
    for i,tweet in enumerate(df1['Tweets']):
      sentence = flair.data.Sentence(tweet)
      sentiment_model.predict(sentence)
      probs.append(sentence.labels[0].score)
      sentiments.append(sentence.labels[0].value)
    New_Sentiment_Class.append(stats.mode(sentiments).mode[0])
    print(index)


```
